```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mimax_normalization(element_array):
    min_val=element_array.min()
    max_val=element_array.max()
    
    array_norm=(element_array-min_val)/(max_val-min_val)
    
    return array_norm
```

```python
def zscore_normalization(element_array):
    mean_val=element_array.mean()
    std_val=element_array.std()
    
    array_norm=(element_array-mean_val)/std_val
    
    return array_norm
```

```python
x=np.random.randint(0,100,100).astype(float)
y=np.random.randint(0,20,100).astype(float)

plt.plot(x,y,'bo')
plt.axis([0, 150, 0, 30])
```

```python

xminmax=mimax_normalization(x)
yminmax=mimax_normalization(y)

plt.plot(xminmax,yminmax,'bo')
plt.axis([0, 1.5, 0, 1.5])
```

```python
xzscore=zscore_normalization(x)
yzscore=zscore_normalization(y)

xnorm=zscore_normalization(xminmax)
ynorm=zscore_normalization(yminmax)


print(np.sum(xnorm-xzscore))
fig=plt.figure(figsize=(18, 4), dpi= 80, facecolor='w', edgecolor='k')

ax1 = plt.subplot(121)
plt.plot(xzscore,yzscore,'bo')

ax2 = plt.subplot(122)
plt.plot(xnorm,ynorm,'bo')
```
